```python
feature_matrix = mat['features'][0][3][2]

#iterate through the rows of feature_matrix and collect the indices where the columns 3,5,12,16 are not zero
indices = []
for i in range(feature_matrix.shape[0]):
    if feature_matrix[i][3] != 0 or feature_matrix[i][5] != 0 or feature_matrix[i][12] != 0 or feature_matrix[i][16] != 0:
        indices.append(i)
indices
```

```python
timings =[]
for i in range(len(data['words'][0])):
    print(data['words'][0][i][1][0][0])
    #store in a list
    timings.append(data['words'][0][i][1][0][0])
```

```python
#create a map for each entry in timings and each index in timeblock
#iterate through timings and find the time block that it belongs to
# if timing[i] is between time_blocks[j][0] and time_blocks[j+1][0] then map the value to the value j
#store the mapping in a dictionary
time_block_map = {}
for i in range(len(timings)):
    for j in range(len(time_blocks)):
        if timings[i] >= time_blocks[j][0] and timings[i] < time_blocks[j+1][0]:
            time_block_map[i] = j
            break





    
```
