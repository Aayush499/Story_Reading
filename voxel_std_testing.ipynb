```python
tr_to_words_mapping = {}
cntr = 0

for i in range( len(subject_1['time'])):
    if cntr < 5176 and subject_1['words'][0][cntr][1] >= subject_1['time'][i][0] and subject_1['words'][0][cntr][1] < (subject_1['time'][i][0]+2):
        tr_to_words_mapping[i] = subject_1['words'][0][cntr][0][0][0][0], subject_1['words'][0][cntr+1][0][0][0][0], subject_1['words'][0][cntr+2][0][0][0][0], subject_1['words'][0][cntr+3][0][0][0][0]
        cntr += 4
    else:
        tr_to_words_mapping[i] = None

 
```

```python
tr_to_words_mapping
```

```python
data = subject_1['data']
```

```python
#Print out the words in index 400 to 700 in tr_to_words_mapping
for i in range(350, 700):
    print(tr_to_words_mapping[i])

```

```python
num_clusters =20

from sklearn.cluster import KMeans


#cluster the data rows

kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(data)

```

```python
k_means_labels = kmeans.labels_
```

```python
for i in k_means_labels:
    print(i)
```

```python
#get the indices of the cluster points labelled 14, 18 , 19 and 1

indices_14 = [i for i, x in enumerate(k_means_labels) if x == 14]
indices_18 = [i for i, x in enumerate(k_means_labels) if x == 18]
indices_19 = [i for i, x in enumerate(k_means_labels) if x == 19]
indices_1 = [i for i, x in enumerate(k_means_labels) if x == 1]


```

```python
#union of the indices
union = list(set(indices_14) | set(indices_18) | set(indices_19) | set(indices_1))

union = sorted(union)
for i in union:
    print(tr_to_words_mapping[i])
```
